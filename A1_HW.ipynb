### 8. Быстрое умножение

```python
import time
from matplotlib import pyplot as plt
```

Определим функции

```python
def NumpyTime(n, it = 10):   
    timer = []
    for i in range(it):
        M = np.random.random(size=(n, n))
        N = np.random.random(size=(n, n))
        t1 = time.time()
        M @ N
        t2 = time.time()
        timer.append(t2 - t1)
    return np.mean(timer)

def SmartForTime(n, it = 10):   
    timer = []
    for k in range(it):
        M = np.random.random(size=(n, n))
        N = np.random.random(size=(n, n))
        P = np.zeros((n,n))
        t1 = time.time()
        for i in range(n):
            for j in range(n):
                P[i,j] = M[i,:]@N[:,j]
        t2 = time.time()
        timer.append(t2 - t1)
    return np.mean(timer)

def DumbForTime(n, it = 10):   
    timer = []
    for k in range(it):
        M = np.random.random(size=(n, n))
        N = np.random.random(size=(n, n))
        P = np.zeros((n,n))
        t1 = time.time()
        for i in range(n):
            for j in range(n):
                sum = 0
                for l in range(n):
                    sum += M[i,l] * N[l,j]
                P[i,j] = sum
        t2 = time.time()
        timer.append(t2 - t1)
    return np.mean(timer)
```

Подберём параменры выборки:

```python
print('\n np_time = ', NumpyTime(100, 1000),'\n sfor_time = ',  SmartForTime(100, 1),'\n dfor_time = ',  DumbForTime(100, 1))
```

Соберём данные:

```python
n_arr = np.arange(10, 100)
np_data = np.vectorize(lambda x: NumpyTime(x, 10))(n_arr)
sfor_data = np.vectorize(lambda x: SmartForTime(x, 1))(n_arr)
dfor_data = np.vectorize(lambda x: DumbForTime(x, 1))(n_arr)
np_data, sfor_data, dfor_data
```

Построим:

```python
plt.figure(figsize=(15, 10))
plt.title('Время выполнения')
plt.scatter(n_arr, np_data, c='r', label='NumPy', marker = '1')
plt.scatter(n_arr, sfor_data, c='b', label='Smart "for"-method', marker = '2')
plt.scatter(n_arr, dfor_data, c='g', label='Straightforward "for"-method', marker = '3')
plt.xlabel('Размер матрицы')
plt.ylabel('Время выполнения, c')
plt.yscale('linear')
plt.grid(True)
plt.legend()
plt.show()
```

```python
n_arr_ext = np.arange(10, 400)
np_data_ext = np.vectorize(lambda x: NumpyTime(x, 10))(n_arr_ext)
```

```python
plt.figure(figsize=(15, 10))
plt.title('Время выполнения')
plt.scatter(n_arr_ext, np_data_ext, c='r', label='NumPy', marker = '1')
plt.scatter(n_arr, sfor_data, c='b', label='Smart "for"-method', marker = '2')
plt.xlabel('Размер матрицы')
plt.ylabel('Время выполнения, c')
plt.yscale('linear')
plt.grid(True)
plt.ylim(0, 0.015)
plt.legend()
plt.show()
```

```python
from scipy import special
```

```python
x = np.linspace(-10, 10, 1000)
ai, aip, bi, bip = special.airy(x)

fig, subplots = plt.subplots(2, 2, figsize=(15, 10))
subplots[0][0].plot(x, ai, 'r', label='Ai(x)')
subplots[0][0].grid()
subplots[0][1].plot(x, aip, 'g--', label='Ai derivative(x)')
subplots[0][1].grid()
subplots[1][0].plot(x[900:1000], bi[900:1000], 'b', label='Bi(x)')
subplots[1][0].grid()
subplots[1][1].plot(x[900:1000], bip[900:1000], 'c--', label='Bi derivative(x)')
subplots[1][1].grid()
plt.show()
```

### Вигнеровский полукруг

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 3000
```

```python
Mat = np.random.normal(size=(N, N))
Mat += Mat.T
```

```python
eig = np.linalg.eigvalsh(Mat)
```

```python
plt.figure(figsize=(8,5))
plt.hist(eig, bins = 120)
plt.show
```

```python
a, b = np.histogram(eig, bins = 120)
b = (b[1:] +b[:-1])/2 # по данным краям бинов получаем центры
plt.figure(figsize=(8,5))
plt.scatter(b,a)
plt.show()
```

```python
n = 500
bin_num = 50
iterate = 50
eigs = np.empty(shape=(1, bin_num))
bins = np.empty(shape=(1, bin_num))

for i in range(iterate):
    M = np.random.normal(size=(n, n))
    M += M.T
    ei = np.linalg.eigvalsh(M)
    a, b = np.histogram(ei, bins = bin_num)
    b = (b[1:] +b[:-1])/2 
    eigs = np.concatenate((eigs, a.reshape(1,-1)), 0)
    bins = np.concatenate((bins, b.reshape(1,-1)), 0)
    
eigval = np.mean(eigs[1:], 0)
bins = np.mean(bins[1:], 0)
yerr = np.std(eigs[1:], 0)
```

```python
plt.figure(figsize=(8,5))
plt.errorbar(bins, eigval, yerr = yerr, marker = '.')
plt.title("Вигнеровский полукруг")
plt.grid(alpha=0.5)
plt.show()
```
