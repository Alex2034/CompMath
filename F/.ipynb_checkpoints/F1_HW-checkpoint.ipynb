```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Monte_Carlo(m, n, k):
  N_plus, N_minus, N_0 = 0, 0, 0
  N, h, s, sc = 10**(k+1), 10**m, 1, 10**(-n)
  mu_0, x_0 = 0, 0
  distribution=np.zeros(100)

  for i in range(N):
    end=0
    x=x_0
    mu=mu_0
    
    while(end==0):
      lam=-1/s*np.log(np.random.rand())
      x=x+lam*mu
      if (x>h):
        N_plus+=1
        end=1
      if (x<0):
        N_minus+=1
        end=1
      if ((x<=h)and(x>=0)):
        if (np.random.randn()<(sc/s)):
          end=1
          N_0+=1
      mu=2*np.random.rand()-1
    

    if ((x<=h)and(x>=0)):
      distribution[int(x/h*100)]+=1
  return N_plus/N, N_minus/N, N_0/N, distribution
```

```python
fig, axs = plt.subplots(4, 1, figsize = (12,12))

data = [[1,2,3],[0.05,1,3],[0.5,1,3],[1,0.01,3]];

for i in range(0,4):
    pt,pr,pa,distr=Monte_Carlo(*data[i])
    ax = axs.flatten()[i]
    ax.plot(np.linspace(0,10,100), distr/(10**(4)*pa), 'r.')
    ax.set_title("m="+str(data[i][0])+", n="+str(data[i][1])+", k="+str(data[i][2])+"\n pt="+str(pt)+"  pr="+str(pr)+"  pa="+str(pa))
```
